```
import numpy as np
import matplotlib.pyplot as plt
import IPython
from sympy import *
data    = np.genfromtxt("data-nonlinear.txt", delimiter=',')

pointX  = data[:, 0]
pointY  = data[:, 1]
label   = data[:, 2]

print(pointX)
print(pointY)
pointX0 = pointX[label == 0]
pointY0 = pointY[label == 0]

pointX1 = pointX[label == 1]
pointY1 = pointY[label == 1]

plt.figure()
plt.scatter(pointX0, pointY0, c='b')
plt.scatter(pointX1, pointY1, c='r')
plt.tight_layout()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

def setup_typeset():
  """MathJax initialization for the current cell.
  
  This installs and configures MathJax for the current output.
  """
  IPython.display.display(IPython.display.HTML('''
      <script src="https://www.gstatic.com/external_hosted/mathjax/latest/MathJax.js?config=TeX-AMS_HTML-full,Safe&delayStartupUntil=configured"></script>
      <script>
        (() => {
          const mathjax = window.MathJax;
          mathjax.Hub.Config({
          'tex2jax': {
            'inlineMath': [['$', '$'], ['\\(', '\\)']],
            'displayMath': [['$$', '$$'], ['\\[', '\\]']],
            'processEscapes': true,
            'processEnvironments': true,
            'skipTags': ['script', 'noscript', 'style', 'textarea', 'code'],
            'displayAlign': 'center',
          },
          'HTML-CSS': {
            'styles': {'.MathJax_Display': {'margin': 0}},
            'linebreaks': {'automatic': true},
            // Disable to prevent OTF font loading, which aren't part of our
            // distribution.
            'imageFont': null,
          },
          'messageStyle': 'none'
        });
        mathjax.Hub.Configured();
      })();
      </script>
      '''))
  
setup_typeset()

x, y, z = symbols('x y z')
a,b,g = symbols('a,b,g')
theta = symbols('theta')
Eq(theta(x**2/a**2+y**2/b**2), g)

theta0 = 0
theta1 = 0
theta2 = 0
theta3 = 0
theta4 = 0
theta5 = 0
theta6 = 0
theta7 = 0
theta8 = 0
theta9 = 0
theta10 = 0
theta11 = 0
theta12 = 0
theta13 = 0
theta14 = 0
theta15 = 0


k=16
alpha = 0.005
i=0
cnt = 0
m=len(pointX)
#energy values
j = np.empty(0)
jx = np.empty(0)
accuracy_arr = np.empty(0)

def g_func():
  return theta0 + theta1*(pointX) + theta2*(pointY)+ theta3*(pointX*pointY)+ theta4*(pointX**2) + theta5*(pointY**2)+ theta6*((pointX**2)*(pointY**2))+theta7*((pointX**2)*(pointY**4))+theta8*((pointX**4)*(pointY**2))+theta9*((pointX**4)*(pointY**4))+ theta10*(pointX**4)+theta11*(pointY**4)+theta12*(pointX**6)+theta13*(pointY**6)+ theta14*((pointX**6)*(pointY**4))+theta15*((pointX*6)*(pointY**6))

def objective_func():
  sigma = 1/(1+np.exp(-1*g_func()))
  j = (1/m)*np.sum((-1*label)*np.log10(sigma)-(1-label)*np.log10(1-sigma))
  return j
  
print(pointX)
print(pointY)
while True:
  cost = objective_func()
  sigma = 1/(1+np.exp(-1*g_func()))  
  theta0 = theta0 - (alpha/m)*np.sum(sigma-label)
  theta1 = theta1 - (alpha/m)*np.sum((sigma-label)*pointX)
  theta2 = theta2 - (alpha/m)*np.sum((sigma-label)*pointY)
  theta3 = theta3 - (alpha/m)*np.sum((sigma-label)*pointX*pointY)
  theta4 = theta4 - (alpha/m)*np.sum((sigma-label)*(pointX**2))
  theta5 = theta5 - (alpha/m)*np.sum((sigma-label)*(pointY**2))
  theta6 = theta6 - (alpha/m)*np.sum((sigma-label)*((pointX**2)*(pointY**2)))
  theta7 = theta7 - (alpha/m)*np.sum((sigma-label)*((pointX**2)*(pointY**4)))
  theta8 = theta8 - (alpha/m)*np.sum((sigma-label)*((pointX**4)*(pointY**2)))
  theta9 = theta9 - (alpha/m)*np.sum((sigma-label)*((pointX**4)*(pointY**4)))
  theta10 = theta10 - (alpha/m)*np.sum((sigma-label)*((pointX**4)))
  theta11 = theta11 - (alpha/m)*np.sum((sigma-label)*((pointY**4)))
  theta12 = theta12 - (alpha/m)*np.sum((sigma-label)*((pointX**6)))
  theta13 = theta13 - (alpha/m)*np.sum((sigma-label)*((pointY**6)))
  theta14 = theta14 - (alpha/m)*np.sum((sigma-label)*((pointX**6)*(pointY**4)))
  theta15 = theta15 - (alpha/m)*np.sum((sigma-label)*((pointX**6)*(pointY**6)))

  sigma = 1/(1+np.exp(-1*g_func()))
  accuracy = 0
  for i in range(len(pointX)):
    if sigma[i]>=(1/2):
      if label[i] ==1:
        accuracy += 1
    elif sigma[i]>0:
      if label[i] == 0:
        accuracy += 1
  accuracy /= len(pointX)
  accuracy_arr = np.append(accuracy_arr, accuracy)
  cost2 = objective_func()
  #append energy values, model parameters
  j = np.append(j, cost2)
  print(cnt)
  #append i for x axis
  jx = np.append(jx, cnt)
  cnt+=1
  if cnt>500000:
    print(cnt)
    print(cost)
    break



#display  model parameters
plt.plot(jx,j,'r-')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Model Parameters')
plt.xlim(0, 1000)
plt.ylim(0, 10)
plt.show()
```
